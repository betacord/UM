# Lab 1 - wprowadzenie do uczenia maszynowego

Metody uczenia maszynowego są znane od lat 50. XX w. Prace nad nimi zostały zapoczątkowane w celu opracowania metod umożliwiających automatyczne rozwiązywanie skomplikowanych problemów. Zaimplementowane algorytmy uczenia maszynowego są dostępne w większości otaczających nas rzeczy, takich jak: smartfon i dostępny w nim asystent głosowy, czy dopasowane reklamy w wyszukiwarce internetowej.

Uczenie maszynowe (ang. machine learning) to dziedzina informatyki zajmująca się metodami automatycznego rozwiązywania skomplikowanych problemów, które są trudne do rozwiązania przy użyciu klasycznych metod programistycznych.

Klasyczne podejście programistyczne polega na zaprojektowaniu przez programistę algorytmu realizującego dany cel, oraz zaimplementowania go w wybranym języku programowania w taki sposób, aby zadanie było automatycznie realizowane przez komputer. Obecnie zdecydowana większość rozwiązań informatycznych jest stworzona w oparciu o takie podejście. Przykład może stanowić obliczanie kwoty zamówienia. Najpierw należy ustalić algorytm, który polega na pomnożeniu ceny każdego z przedmiotów z zamawianą ilością, a następnie zsumować wszystko ze sobą. Następnie należy zaimplementować algorytm w wybranym języku programowania. Cały proces składał się z dwóch kroków, wykonywanych całkowicie przez człowieka.

W celu lepszego zrozumienia idei uczenia maszynowego warto przytoczyć przykład komputerowego rozpoznawania obrazów polegający na rozróżnieniu jakie zwierze jest przedstawione na danym obrazie. Dla ułatwienia zdjęcia mogą przedstawiać jedynie psa, lub kota posiadając jednocześnie etykietę informaującą jakie zdjęcie jest przedstawione. Na etapie ustalania algorytmu programista musi opracować zbiór reguł decyzyjnych, które pewien podzbiór pikseli ze zdjęcia przekształcą do postaci przedstawionej na etykiecie, a następnie zaimplementuje algorytm w wybranym języku programowania.

Uczenie maszynowe pozwala rozwiązać taki problem w sposób automatyczny. Algorytmy uczące przystosowane są do rozwiązywania problemów generycznie, a ich celem jest wytrenowanie wewnętrznych reguł decyzyjnych w celu wykonania możliwie jak najbardziej dokładnych prognoz. Oznacza to, że na wejściu takiego algorytmu należy przekazać zbiór uczący, który składa się z konkretnych przykładów rozwiązanych przez człowieka. Zbiór przykładów uczących mogą stanowić zdjęcia przedstawiające psy i koty opatrzone etykietą informującą jakie zwierzę się tam znajduje. Zatem celem algorytmów uczenia maszynowego jest dostosowanie wewnętrznych reguł decyzyjnych w taki sposób, aby pewien zbiór przekształcić na inny zbiór. Jako przykład można przytoczyć przekształcanie zbioru zdjęć na zbiór etykiet o treści "pies", lub "kot".

## System informacyjny

Niech $S = (U, A)$ będzie systemem informacyjnym

gdzie:
- $ U = \{x_1, x_2, ..., x_n\} $ oznacza niepusty skończony obiektów, zwany uniwersum
- $ A = \{a_1, a_2, ..., a_n\} $ oznacza skończony niepusty zbiór atrybutów

Każdy atrybut $a \in A$ powstaje z transformacji $a: U \rightarrow V_a$, gdzie na podstawie zbioru obiektów powstaje wektor wartości atrybutu.

Dla dowolnego atrybutu $a$ oraz jego wartości $v$ deskryptor jest wyrażony następującą formułą: $(a = v)$, którą w zbiorze obiektów można przedstawić następująco:
$ [[(a = v)]] = \{u \in U: a(u) = v\} $

Intuicyjnie deskryptor można rozumieć jako wartość atrybutu pewnego obiektu. W poniższej przedstawiono przykładowy system informacyjny. W obiekcie numer 3 można znaleźć siłę wiatru w postaci deskryptora (wiatr = umiarkowany).

## System decyzyjny

System decyzyjny to uporządkowana trójka $(U, A, d)$

gdzie:
- $U$ oznacza skończony zbiór obiektów
- $A$ oznacza skończony zbiór atrybutów
- $d$ oznacza atrybut decyzyjny, czyli skończony wektor decyzji dla obiektów

Zbiór $A$ zawiera atrybuty warunkowe, wyznaczane empirycznie przez twórcę systemu. Fakt skończoności zbioru $A$ może wpływać na rozróżnialność obiektów. Atrybut decyzyjny (wektor decyzji dla obiektów) jest odwzorowaniem konkretnych procesów zachodzących w realnym świecie przy pomocy eksperta. W języku deskryptorów obiekty $u \in U$ można przedstawić w postaci wektorów informacyjnych:
$ Inf_A(u) = \{(a, a(u)): a \in A\} $

gdzie:
- $a$ oznacza atrybut pochodzący ze zbioru $A$
- $u$ oznacza obiekt pochodzący ze zbioru $U$
- $a(u)$ oznacza wartość atrybutu $a$ dla obiektu $u$

## Uczenie nadzorowane

Uczenie nadzorowane polega na znalezieniu najlepszej konfiguracji parametrów modelu predykcyjnego w taki sposób, aby najlepiej generalizował wyniki na obiektach, które nie były przetwarzane na etapie uczenia.
Ocena generalizacji modelu polega na porównaniu prognoz modelu i prognoz przypisanych obiektom w atrybucie decyzyjnym przy użyciu funkcji błędu.

## Uczenie nienadzorowane
W uczeniu nienadzorowanym (ang. unsupervised learning) dany jest zbiór przykładów uczących bez przydzielonej decyzji. Na podstawie dużej liczby obiektów algorytmy uczenia nienadzorowanego mogą dokonać ich grupowania, w oparciu o podobieństwo geometryczne. Powstałe w ten spobób grupy noszą nazwę klastrów.

Klastry nie zapewniają jednak odpowiedzi, czym są obiekty się w nich znajdujące. Klastry odpowiadają jedynie na pytania, które obiekty są do siebie podobne na podstawie wskazanych atrybutów.

## Typowy potok uczenia maszynowego

1. Określenie problemu do rozwiązania.
2. Zgromadzenie danych.
   1. Zdefiniowanie odpowiedniej przestrzeni.
   2. Dokumentacja danych.
   3. Anonimizacja.
3. Podział danych.
   1. Wydzielenie rozłącznych podzbiorów: treningowy oraz testowy.
   2. Proporcje dostosowane do rozmiaru pełnego zbioru danych. 
   3. Wszelkie dalsze kroki dokonywane tylko na podzbiorze treningowym.
4. Przetwarzanie i oczyszczanie danych.
   1. Uzupełnianie brakujących wartości.
   2. Kodowanie numeryczne wartości tekstowych.
   3. Redukcja asymptot.
   4. Skalowanie, standaryzacja i normalizacja wartości.
   5. Redukcja wymiarowości.
5. Trening modelu.
   1. Dopasowanie architektury modelu.
   2. Optymalizacja hiperparametrów.
   3. Analiza miar.
6. Uformowanie finalnego potoku.
